```python
game_df = tup[0]
pbp_df=tup[1]
```

```python
team_ids = pd.concat(
    [game_df[['home_id', 'home_team']].rename(columns={'home_id':'id','home_team':'team'}), game_df[['away_id', 'away_team']].rename(columns={'away_id':'id','away_team':'team'})]
)
team_ids.drop_duplicates(inplace=True)
```

```python
game_df[game_df.home_id == "333"]
```

```python
team_ids[team_ids.team.str.contains("Maryland")]
```

```python
def get_team_games(team_name : str):
    id = team_ids[team_ids.team == team_name].id.item()
    return game_df[game_df.home_id.str.match(id) | game_df.away_id.str.match(id)]
```
